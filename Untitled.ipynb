```python
import nltk
import bs4
def summary_articles():
    try:
        url=input("enter url to check :")
        import requests
        html=requests.get(url)
        soup=bs4.BeautifulSoup(html.text,"lxml")
        t=[]
        for item in soup.select('p'):
            t.append(item.text)
        if len(t)==0:
            raise Exception
        stop=nltk.corpus.stopwords.words("english") 
        tokens=nltk.word_tokenize(str(t))
        info=[]
        for word in tokens:
            if word.isalpha(): 
                info.append(word.lower())
        relevant=[]
        for w in info:
            if w not in stop:
                relevant.append(w) 
        fd=nltk.FreqDist(relevant)
        fd.plot(20,cumulative=False)
    except:
        print("please check for permissions on the given url")
```

```python
summary_articles()
```
